# Array Attributes :

```python
a1 = np.arange(10, dtype=int)
a2 = np.arange(10,dtype=float).reshape(2,5)
a3 = np.arange(8).reshape(2,2,2)
a3
```

```python
#  ndim (used for checking array dimension) :

a3.ndim
```

```python
# shape : (tells present total number of items in given array)

print(a1.shape)

print(a2.shape)

print(a3.shape)
```

```python
# size
print(a1.size)
```

```python
# itemsize : space occupied in memory

print(a3.itemsize)
```

```python
# dtype:
print(a1.dtype)
print(a2.dtype)
print(a3.dtype)
```

* Changing Datatype :

```python
# astype : (Used for saving space in memory by changing it's datatype)
a3.astype(np.int32)
```

# **Array Operations :**

```python
# Scalar operations :

#  (Arithmetic operation):

a3 + 2
```

```python
# relational operations: (Comparision)

print(a3 == 2)
print(a3 >1)
```

```python
#  Vector Operations : (when you have to work with more than 1 numpy array then use vector operations)

a4 = np.arange(10).reshape(2,5)
print(a4)
a2 + a4
```

# **Numpy array functions :**

```python
a1
```

```python
# max/min/sum/prod:

print("Max:",np.max(a1))
print("Min:",np.min(a1))
print("Sum:",np.sum(a1))
print("Product:",np.prod(a2))
```

```python
# max/min/sum/prod:
# 0 -> column and 1 -> row

a5 = np.random.random((3,3))
print(a5)
print("-"*25)
print(np.sum(a5,axis=0))
```

```python
# mean/median/std/var

print(np.mean(a5))
print(np.mean(a5,axis=1))
print(np.median(a5))
print(np.std(a5))
print(np.var(a5))
```

```python
# trigonometric functions:
np.sin(a1)
```

```python
# dot product:

a2 = np.arange(12).reshape(3,4)
print(a2)

a3 = np.arange(12,24).reshape(4,3)
print(a3)

np.dot(a2,a3)
```

```python
# log and exponents:
print(np.log(a3))
print(np.exp(a3))
```

```python
# round/floor/ceil

print("Round:",np.round(np.random.random((2,3))*100))
print("Floor:",np.floor(np.random.random((2,3))*100))
print("Ceil:",np.ceil(np.random.random((2,3))*100))
```

# **Indexing and slicing:**

```python
a1 = np.arange(10, dtype=int)
a2 = np.arange(12,dtype=int).reshape(4,3)
a3 = np.arange(8).reshape(2,2,2)
a3
```

```python
a1[0]
a1[-1]
```

```python
a2
```

```python
# find 7 at which index?

# num_array[row,column]
print(a2[2,1])

print(a2[2:])
```

# **Iterating :**

```python
# 1D array
for i in a1:
  print(i)
```

```python
# 2D array:
for i in a2:
  print(i)
```

```python
# 3D array:
for i in a3:
  print(i)

# for iterating each item from any 2D to nD array:

for i in np.nditer(a3):
  print(i)
```
